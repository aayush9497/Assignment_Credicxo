```python
X=np.array(shuffled_data)
Y=np.array(labels)
Y=Y.reshape(Y.shape[0],1)
```

The dataset is unbalanced i.e. Majority class=0 and Minority class=1

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, Y, test_size = 0.2, shuffle=False)
```

```python
xTrain.shape
```

```python
model=keras.models.Sequential()
```

```python
model.add(keras.layers.Dense(units=xTrain.shape[1],activation=keras.activations.relu,
                                kernel_initializer=keras.initializers.he_uniform(),
                                kernel_regularizer=keras.regularizers.l2(),
                                activity_regularizer=keras.layers.BatchNormalization(),
                                input_shape=(xTrain.shape[1],)))
```

```python
model.add(keras.layers.Dense(units=xTrain.shape[1],activation=keras.activations.relu,
                                kernel_initializer=keras.initializers.he_uniform(),
                                kernel_regularizer=keras.regularizers.l2(),
                                activity_regularizer=keras.layers.BatchNormalization(),
                                
                            ))
```

```python
model.add(keras.layers.Dense(units=xTrain.shape[1],activation=keras.activations.relu,
                                kernel_initializer=keras.initializers.he_uniform(),
                                kernel_regularizer=keras.regularizers.l2(),
                                activity_regularizer=keras.layers.BatchNormalization()
                                ))
```

```python
model.add(keras.layers.Dense(units=xTrain.shape[1],activation=keras.activations.relu,
                                kernel_initializer=keras.initializers.he_uniform(),
                                kernel_regularizer=keras.regularizers.l2(),
                                activity_regularizer=keras.layers.BatchNormalization()
                                ))
```

```python
model.add(keras.layers.Dense(units=2,activation=keras.activations.softmax))
```

```python
model.summary()
```

```python
def recall_m(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + K.epsilon())
        return recall

def precision_m(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + K.epsilon())
        return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
y_train=keras.utils.to_categorical(yTrain)
y_test=keras.utils.to_categorical(yTest)
```

```python
model.compile(optimizer=keras.optimizers.Adam(),loss=keras.losses.categorical_crossentropy,metrics=['accuracy',f1_m,precision_m,recall_m])
```

```python
weights={0: 0.2,
        1: 0.8}
```

```python
m=model.fit(x=xTrain,y=y_train,batch_size=91,epochs=30,class_weight=weights,validation_split=0.155)
```

```python
res=model.predict(xTest)
```

```python
classes_predicted=np.argmax(res,axis=1)
```

```python
c=classes_predicted.reshape(classes_predicted.shape[0],1)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(yTest,c,target_names=['0','1']))
```
